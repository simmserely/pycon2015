### List the supporting roles (having n=2) played by Cary Grant in the 1940s, in order by year.

```python
carygrant_supporting = cast[(cast.name == 'Cary Grant') & (cast.n == 2) & (cast.year <= 1940) & (cast.year < 1950)]
carygrant_supporting.sort_values(by='year')
```

### List the leading roles that Cary Grant played in the 1940s in order by year.

```python
carygrant_leading = cast[(cast.name == 'Cary Grant') &(cast.n == 1) & (cast.year.between(1940, 1950, inclusive=False))]
carygrant_leading.sort_values(by='year')
```
